```python
# webscrap pricing details of houses in kerala from makaan.com

import csv
import bs4
import requests

header = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"
}

link = "https://www.makaan.com/kerala/buy-property-in-kerala-state"
num_pages = 362
data = []

for n in range(num_pages):
    print(f"collecting page {n+1} details...")
    
    link = f"{link}?page={n}"
    response = requests.get(link, headers=header, timeout=15)
    soup = bs4.BeautifulSoup(response.text, 'lxml')
    read = soup.select(".cardholder")

    for i, d in enumerate(read):
        collected = {}

        title = d.find_all("div", {"class":"title-line"})
        _title = title[0].getText() if len(title) else None
        if _title == 'Residential Plot':
            continue
        collected['title'] = _title
        
        city = d.find_all("span", {"class":"cityName"})
        collected['city'] = city[0].getText() if len(city) else None  

        location = d.find_all("span", {"class":"locName"})
        collected['location'] = location[0].getText().split(',', 1)[0] if len(location) else None
            
        price = d.find_all("td", {"class":"price"})
        collected['price'] = price[0].getText() if len(price) else None

        area = d.find_all("td", {"class":"size"})
        collected['area'] = area[0].getText() if len(area) else None

        construction_status = d.find_all("tr", {"class":"hcol w44"})
        collected['construction_status'] = construction_status[0].getText() if len(construction_status) else None

        bathrooms = d.find_all("li", {"class":"keypoint", "title":"Bathrooms"})
        collected['bathrooms'] = bathrooms[0].getText() if len(bathrooms) else None

        age = d.find_all("li", {"class":"keypoint", "title":"old"})
        collected['age'] = age[0].getText() if len(age) else None    

        data.append(collected)

keys = data[0].keys()

with open('kerala_house_data.csv', 'w', newline='') as output_file:
    writer = csv.DictWriter(output_file, keys)
    writer.writeheader()
    writer.writerows(data)
```
